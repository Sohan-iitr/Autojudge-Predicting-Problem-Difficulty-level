```python
# 1. Convert 'problem_score' to numbers. 
# 'coerce' means if it finds a typo (like "N/A"), it turns it into NaN (Not a Number).
df['problem_score'] = pd.to_numeric(df['problem_score'], errors='coerce')

# 2. Drop rows where the score is missing/invalid
# We call this new clean version 'df_clean'
df_clean = df.dropna(subset=['problem_score'])

# 3. Also combine the text columns now (for the advanced analytics)
df_clean['full_text'] = (df_clean['description'].fillna('') + " " + 
                         df_clean['input_description'].fillna('') + " " + 
                         df_clean['output_description'].fillna(''))

print(f"Original shape: {df.shape}")
print(f"Clean shape: {df_clean.shape}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
from sklearn.feature_extraction.text import CountVectorizer

# ==========================================
# 1. MATH & SYMBOL ANALYSIS
# ==========================================
# Function to count math-like symbols
def count_math_symbols(text):
    # Counts symbols like +, -, *, /, %, ^, =, <, >
    return len(re.findall(r'[+\-*/%^=<>_]', str(text)))

# Function to count total numbers (constraints often appear as numbers)
def count_numbers(text):
    return len(re.findall(r'\d+', str(text)))

df_clean['math_symbol_count'] = df_clean['full_text'].apply(count_math_symbols)
df_clean['number_count'] = df_clean['full_text'].apply(count_numbers)

# Plotting Math Density
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='problem_class', y='math_symbol_count', data=df_clean, order=['easy', 'medium', 'hard'])
plt.title('Frequency of Math Symbols by Difficulty')
plt.grid(axis='y', linestyle=':', alpha=0.5)

plt.subplot(1, 2, 2)
sns.boxplot(x='problem_class', y='number_count', data=df_clean, order=['easy', 'medium', 'hard'])
plt.title('Frequency of Numbers (Constraints) by Difficulty')
plt.grid(axis='y', linestyle=':', alpha=0.5)

plt.tight_layout()
plt.show()

# ==========================================
# 2. BIGRAM ANALYSIS (Top Phrases)
# ==========================================
def get_top_bigrams(text_series, n=10):
    vec = CountVectorizer(ngram_range=(2, 2), stop_words='english')
    bag_of_words = vec.fit_transform(text_series.dropna())
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)
    return words_freq[:n]

# Visualize Top Bigrams for "Hard" Problems
hard_text = df_clean[df_clean['problem_class'] == 'hard']['full_text']

if not hard_text.empty:
    top_hard_bigrams = get_top_bigrams(hard_text)
    x_val = [x[0] for x in top_hard_bigrams]
    y_val = [x[1] for x in top_hard_bigrams]

    plt.figure(figsize=(10, 6))
    sns.barplot(x=y_val, y=x_val, palette='viridis')
    plt.title('Top 10 Key Phrases (Bigrams) in Hard Problems')
    plt.xlabel('Frequency')
    plt.grid(axis='x', linestyle=':', alpha=0.5)
    plt.show()
```
